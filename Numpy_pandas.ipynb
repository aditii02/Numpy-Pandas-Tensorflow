```python
import numpy as np

array_np = np.array([1,2,3,4],dtype = "int")
print(repr(array_np)) #represented as a array


#copy
a = np.array([1,2,3,4],dtype = "int") #bydefault array type = float32
b = a.copy()
c = a
b[1] = 7
c[2] = 9 # will change the array a as well
print(repr(a))
print(repr(b))

#casting
a = a.astype(np.float32)
print(a)

#NaN value

a = np.array([np.NaN, 1,2,3],dtype = np.float32)

#infinity

a = np.array([np.inf,9,0,8])
b = np.array([-np.inf,7,0,8])
print(a)
print(b)

# range in np array

a = np.arange(5)
b = np.arange(5,10,2)
print(a,b)

# use of linspace. - gives out the array equally dividing between the numbers given, endpoint is included , set it false using endpoint = False

a = np.linspace(1,10,8, endpoint = False)
print("using linspace",a)


# reshape

arr = np.arange(8)
reshap = np.reshape(arr,(4,2))
print("reshap array",reshap)
print("shape of the array", reshap.shape)

reshaped_arr = np.reshape(arr, (-1, 2, 1)) # -1 will take any value to fit the array into (2,1), in this case its 4 arrays of shape (2,1)
print(repr(reshaped_arr))
print('New shape: {}'.format(reshaped_arr.shape))

#transpose

transposed = np.transpose(reshap)
print("after transpose",repr(transposed))

#flattening the array

new_array = arr.flatten()
print("new_array",new_array)

#np zeros and ones

zeros= np.zeros(4)
ones = np.ones((2,3),np.float32)
print(zeros,ones)

#applying maths


a = a/2
print("after maths",a)
d = np.array([1,2,3,4])
f = np.array([[1],[2],[3],[4]])
log = np.log2(d)
print("after log",log)

mat = np.matmul(d,f)
print("after multiplying",mat)



# random in numpy

ar1 = np.random.randint(1000) #generate one number ranging between 0 and 1000
ar2 = np.random.randint(6,high =14) #generate one number between 6 and 14
ar3 = np.random.randint(6, high = 20, size = (2,3)) # generates an array (2,3) and every number is between 6 and 20
print("random array generated", ar1,ar2,ar3)

#set random seed. - the input to be given to the random number generator if not given it takes up the default value, lets us control the outputs of the random generator

np.random.seed(1000)
ar4 = np.random.randint(9) #if high not given , consider it from 0 to 9, if high=90 given then 9 to high value
np.random.seed(98)
ar5 = np.random.randint(8,high=70, size = (3,2))
print(ar5)
print(ar4)

#shuffling the array, shuffle around matrix will only shuffle the rows

ar5 = np.array([9,0,7,8,5])
print(np.random.shuffle(ar5))

#custom sampling
colors = ["red","blue","yellow"]
sample = np.random.choice(colors, size = (2,2),p= [0.3,0.3,0.4])#p is the probability of getting selected, if size = (2,2), then you will arrange 4 colors randomly selected based on probability
print(sample)
```

```python
#group by on dataframe

mlb_df1 = pd.DataFrame({'name': ['john doe', 'al smith', 'sam black', 'john doe'],
                        'pos': ['9','8','2','0'],
                        'year': [2004, 2004, 2008, 2003]})
mlb_df2 = pd.DataFrame({'name': ['john doe', 'al smith', 'jack lee','john doe'],
                        'year': [2000, 2004, 2012,2003],
                        'rbi': [80, 100, 12,90]})
mlb_df3 = pd.concat([mlb_df1,mlb_df2])
print("mlb_df3",mlb_df3)
groups = mlb_df3.groupby('year') #group by year
for n,grp in groups:
  print("n",n)
  print("grp",grp)
group_sum = groups.sum()
print("sum of the groups",group_sum)
print("get a particular group",groups.get_group(2004))
group25 = groups.filter(lambda x: x.name > 2012) # filter by year >2003
print("group 25",group25)


#features in a data frame - Quantitative or categorical

data_frame1 = pd.DataFrame({"T!":[1,2,3],"T@":[4,5,6],"T#":[7,8,9]})
data_sum = data_frame1.sum(axis=1) # row wise sum
print("data_sum",data_sum)
print("multiply with a number",data_frame1.multiply(3)) #multiply each number by 3
df4 = data_frame1.multiply([2,4,0],axis = 0) #column wise multiply
print("df4",df4)


```
